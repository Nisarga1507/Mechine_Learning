```python
import os
import cv2
import numpy as np
import pandas as pd
from pathlib import Path
from PIL import Image
from skimage import morphology, measure, segmentation, filters
from scipy import ndimage as ndi
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# --------------------------
# User config
# --------------------------
img_dir = '/content/drive/MyDrive/Colab Notebooks/Image/img_dir'
ann_dir = '/content/drive/MyDrive/Colab Notebooks/Image/ann_dir'
output_dir = '/content/drive/MyDrive/Colab Notebooks/Image/segmentation_output'
os.makedirs(output_dir, exist_ok=True)

EARLY_STOP_IoU = 0.60  # stop when IoU â‰¥ this

# Expected object size characteristics (tweak for your dataset)
EXPECTED_MIN_AREA = 200    # smallest cyst pixel area you consider real
EXPECTED_MAX_AREA = 200000 # largest area (very large = likely background)
EXPECTED_CENTER_Y_RANGE = (0.15, 0.85)  # fraction of height where lesions usually occur

# --------------------------
# Helpers
# --------------------------
def read_gray(path):
    return np.array(Image.open(path).convert('L'))

def save_uint8(path, arr):
    Image.fromarray(arr.astype(np.uint8)).save(path)

def apply_clahe(img, clip=2.0, tile=(8,8)):
    clahe = cv2.createCLAHE(clipLimit=clip, tileGridSize=tile)
    return clahe.apply(img.astype(np.uint8))

def remove_small_and_holes(bin_mask, min_size=200, min_hole=200):
    b = bin_mask>0
    b = morphology.remove_small_objects(b, min_size=min_size)
    b = morphology.remove_small_holes(b, area_threshold=min_hole)
    return (b.astype(np.uint8) * 255)

def overlay_contours(orig_gray, mask):
    rgb = cv2.cvtColor(orig_gray, cv2.COLOR_GRAY2BGR)
    contours, _ = cv2.findContours((mask>0).astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(rgb, contours, -1, (255,0,0), 1)
    return rgb, len(contours)

def compute_metrics(pred, gt):
    pred_b = (pred>0).astype(np.uint8)
    gt_b = (gt>0).astype(np.uint8)
    inter = np.logical_and(pred_b, gt_b).sum()
    union = np.logical_or(pred_b, gt_b).sum()
    iou = inter/union if union else 0.0
    dice = (2*inter)/(pred_b.sum()+gt_b.sum()) if (pred_b.sum()+gt_b.sum()) else 0.0
    return float(iou), float(dice)

# --------------------------
# Robust cleaning (as before)
# --------------------------
def clean_mask(bin_mask,
               min_size=500,
               area_percentile=25,
               ecc_max=0.98,
               opening_k=5,
               closing_k=9,
               median_ksize=5,
               take_convex_hull=False):
    b = (bin_mask > 0).astype(bool)

    if median_ksize and median_ksize > 1:
        b = cv2.medianBlur((b.astype(np.uint8)*255), median_ksize) > 0

    se_o = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (opening_k, opening_k))
    b = cv2.morphologyEx(b.astype(np.uint8), cv2.MORPH_OPEN, se_o).astype(bool)

    se_c = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (closing_k, closing_k))
    b = cv2.morphologyEx(b.astype(np.uint8), cv2.MORPH_CLOSE, se_c).astype(bool)

    b = morphology.remove_small_objects(b, min_size=50)
    b = morphology.remove_small_holes(b, area_threshold=50)

    labels = measure.label(b, connectivity=2)
    props = measure.regionprops(labels)

    if len(props) == 0:
        return (b.astype(np.uint8) * 255)

    areas = np.array([p.area for p in props])
    thr = max(min_size, int(np.percentile(areas, area_percentile)))
    out = np.zeros_like(labels, dtype=np.uint8)

    for p in props:
        if p.area >= thr and p.eccentricity < ecc_max:
            out[labels == p.label] = 255

    if take_convex_hull and out.sum()>0:
        lbl2 = measure.label(out>0)
        hull_mask = np.zeros_like(out, dtype=np.uint8)
        for region in measure.regionprops(lbl2):
            coords = region.coords
            reg_mask = np.zeros_like(out, dtype=np.uint8)
            reg_mask[coords[:,0], coords[:,1]] = 1
            ch = morphology.convex_hull_image(reg_mask>0)
            hull_mask[ch] = 255
        out = hull_mask

    out_bool = (out>0)
    out_bool = morphology.remove_small_holes(out_bool, area_threshold=100)
    out_bool = morphology.remove_small_objects(out_bool, min_size=max(20, min_size//2))

    return (out_bool.astype(np.uint8) * 255)

# --------------------------
# Smart component selection
# --------------------------
def select_components(mask_u, orig_img,
                      expected_min_area=EXPECTED_MIN_AREA,
                      expected_max_area=EXPECTED_MAX_AREA,
                      center_y_range=EXPECTED_CENTER_Y_RANGE):
    """
    Take a cleaned binary mask and pick components most likely to be true cysts.
    Returns a mask (uint8 0/255) made from the selected components.
    Strategy:
      - remove components touching the image border (likely outside).
      - compute per-component features: area, solidity, mean intensity, centroid.
      - score components by weighted sum of normalized features.
      - keep components with score above threshold (or keep top k).
    """

    h, w = orig_img.shape[:2]
    lbl = measure.label(mask_u>0, connectivity=2)
    props = measure.regionprops(lbl, intensity_image=orig_img)
    if len(props) == 0:
        return np.zeros_like(mask_u, dtype=np.uint8)

    comps = []
    for p in props:
        minr, minc, maxr, maxc = p.bbox
        # check border touching
        touches_border = (minr <= 5) or (minc <= 1) or (maxr >= h-6) or (maxc >= w-6)
        area = p.area
        solidity = getattr(p, "solidity", 0.0)
        mean_int = p.mean_intensity if hasattr(p, "mean_intensity") else 0.0
        cy, cx = p.centroid
        # centrality: how close centroid Y is to expected retinal band area
        cy_norm = cy / h
        center_score = 1.0 if (center_y_range[0] <= cy_norm <= center_y_range[1]) else 0.0
        comps.append({
            "label": p.label,
            "area": area,
            "solidity": solidity,
            "mean_intensity": mean_int,
            "touches_border": touches_border,
            "centroid_y_norm": cy_norm,
            "center_score": center_score,
            "bbox": p.bbox
        })

    # Convert to DataFrame for easy scoring
    dfc = pd.DataFrame(comps)
    if dfc.shape[0] == 0:
        return np.zeros_like(mask_u, dtype=np.uint8)

    # Remove components touching border (these are usually outside artifacts)
    dfc = dfc[~dfc["touches_border"]].copy()
    if dfc.shape[0] == 0:
        # If all touched border, allow them but mark in scoring (less preferred)
        dfc = pd.DataFrame(comps)

    # Normalize features
    # Check if there is variation before normalizing to avoid division by zero
    area_range = dfc["area"].max() - dfc["area"].min()
    dfc["area_n"] = (dfc["area"] - dfc["area"].min()) / (area_range + 1e-8) if area_range > 0 else 0.0

    solidity_range = dfc["solidity"].max() - dfc["solidity"].min()
    dfc["solidity_n"] = (dfc["solidity"] - dfc["solidity"].min()) / (solidity_range + 1e-8) if solidity_range > 0 else 0.0

    mean_int_range = dfc["mean_intensity"].max() - dfc["mean_intensity"].min()
    dfc["mean_int_n"] = (dfc["mean_intensity"] - dfc["mean_intensity"].min()) / (mean_int_range + 1e-8) if mean_int_range > 0 else 0.0

    # Score weights (tweak these)
    w_area = 0.4
    w_solidity = 0.2
    w_mean_int = 0.2
    w_center = 0.2

    dfc["score"] = (w_area * dfc["area_n"] +
                    w_solidity * dfc["solidity_n"] +
                    w_mean_int * dfc["mean_int_n"] +
                    w_center * dfc["center_score"])

    # filter by expected sizes
    dfc = dfc[(dfc["area"] >= expected_min_area) & (dfc["area"] <= expected_max_area)]

    if dfc.shape[0] == 0:
        # nothing fits size filters; relax min area a bit
        dfc = pd.DataFrame(comps)
        dfc = dfc[dfc["area"] >= max(20, expected_min_area//4)]
        if dfc.shape[0] == 0:
            return np.zeros_like(mask_u, dtype=np.uint8)

        # Recalculate normalized features for relaxed filter
        area_range = dfc["area"].max() - dfc["area"].min()
        dfc["area_n"] = (dfc["area"] - dfc["area"].min()) / (area_range + 1e-8) if area_range > 0 else 0.0

        solidity_range = dfc["solidity"].max() - dfc["solidity"].min()
        dfc["solidity_n"] = (dfc["solidity"] - dfc["solidity"].min()) / (solidity_range + 1e-8) if solidity_range > 0 else 0.0

        mean_int_range = dfc["mean_intensity"].max() - dfc["mean_intensity"].min()
        dfc["mean_int_n"] = (dfc["mean_intensity"] - dfc["mean_intensity"].min()) / (mean_int_range + 1e-8) if mean_int_range > 0 else 0.0

        dfc["score"] = (w_area * dfc["area_n"] +
                        w_solidity * dfc["solidity_n"] +
                        w_mean_int * dfc["mean_int_n"] +
                        w_center * dfc["center_score"])

    # Keep components with score above threshold or top-k by score:
    score_thr = 0.25
    selected = dfc[dfc["score"] >= score_thr].sort_values("score", ascending=False)
    if selected.shape[0] == 0:
        # take top-1 if none exceed threshold
        selected = dfc.sort_values("score", ascending=False).head(1)

    # Compose final mask from selected labels
    sel_labels = set(selected["label"].astype(int).tolist())
    final_mask = np.zeros_like(mask_u, dtype=np.uint8)
    for lbl_id in sel_labels:
        final_mask[lbl == lbl_id] = 255

    # Final morphological clean
    final_mask = cv2.morphologyEx(final_mask, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5)))
    final_mask = morphology.remove_small_objects(final_mask>0, min_size=expected_min_area//2)
    final_mask = (final_mask.astype(np.uint8) * 255)

    return final_mask

# --------------------------
# Existing segmentation strategies (unchanged)
# --------------------------
def strat_otsu_clahe(img, invert=True, clahe_clip=2.0, blur_k=3, open_k=3, min_size=200, min_hole=200):
    im = apply_clahe(img, clip=clahe_clip)
    if blur_k and blur_k>1:
        im = cv2.GaussianBlur(im, (blur_k|1, blur_k|1), 0)
    if invert:
        _, th = cv2.threshold(cv2.bitwise_not(im), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    else:
        _, th = cv2.threshold(im, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    th = cv2.morphologyEx(th, cv2.MORPH_OPEN, cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(open_k,open_k)))
    th = cv2.morphologyEx(th, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(open_k,open_k)))
    th = remove_small_and_holes(th, min_size=min_size, min_hole=min_hole)
    return th

def strat_adaptive(img, invert=False, blur_k=3, blockSize=11, C=2, min_size=100, min_hole=100):
    im = img.copy()
    if blur_k and blur_k>1:
        im = cv2.GaussianBlur(im, (blur_k|1, blur_k|1), 0)
    th = cv2.adaptiveThreshold(im,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, blockSize, C)
    if invert:
        th = cv2.bitwise_not(th)
    th = remove_small_and_holes(th, min_size=min_size, min_hole=min_hole)
    return th

def strat_canny_watershed(img, sigma=1.0, canny_low=50, canny_high=150, min_size=100):
    blurred = cv2.GaussianBlur(img, (3,3), sigma)
    edges = cv2.Canny(blurred, canny_low, canny_high)
    filled = ndi.binary_fill_holes(edges==0)
    filled = (filled.astype(np.uint8)*255)
    dist = ndi.distance_transform_edt(filled>0)
    local_max = morphology.h_maxima(dist, 5)
    markers = measure.label(local_max)
    if markers.max() == 0:
        markers = measure.label(dist > (dist.max()*0.2))
    labels = segmentation.watershed(-dist, markers, mask=filled>0)
    out = np.zeros_like(img, dtype=np.uint8)
    out[labels>0] = 255
    out = remove_small_and_holes(out, min_size=min_size, min_hole=min_size)
    return out

# --------------------------
# Candidate generation
# --------------------------
def generate_candidates(img):
    candidates = []
    for inv in [True, False]:
        for clip in [1.5, 2.0, 3.0]:
            for blur in [1,3]:
                m = strat_otsu_clahe(img, invert=inv, clahe_clip=clip, blur_k=blur,
                                     open_k=3, min_size=100, min_hole=100)
                candidates.append(('otsu_clahe', {'invert':inv,'clip':clip,'blur':blur}, m))
    for inv in [True, False]:
        for blur in [1,3]:
            m = strat_adaptive(img, invert=inv, blur_k=blur, blockSize=11, C=2, min_size=50, min_hole=50)
            candidates.append(('adaptive', {'invert':inv,'blur':blur}, m))
    for (low,high) in [(30,100),(50,150),(80,200)]:
        m = strat_canny_watershed(img, sigma=1.0, canny_low=low, canny_high=high, min_size=50)
        candidates.append(('canny_ws', {'low':low,'high':high}, m))
    # extra morphological top-hat candidate (helps reveal dark cysts)
    try:
        se = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (15,15))
        tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, se)
        # threshold top-hat to get bright spots
        _, th = cv2.threshold(tophat, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        candidates.append(('tophat', {'se':15}, th))
    except Exception:
        pass
    return candidates

# --------------------------
# Main loop
# --------------------------
img_dir_p = Path(img_dir)
ann_dir_p = Path(ann_dir)
output_dir_p = Path(output_dir)

rows = []
for img_path in sorted(img_dir_p.glob("*.tif")):
    print("\nProcessing:", img_path.name)
    img = read_gray(img_path)
    print("  img shape/dtype:", img.shape, img.dtype, "min/max:", img.min(), img.max())

    ann_path = ann_dir_p / img_path.name
    gt = None
    if ann_path.exists():
        gt = read_gray(ann_path)
        print("  GT found:", ann_path.name, "shape/dtype:", gt.shape, gt.dtype, "unique (sample):", np.unique(gt)[:10])
        if gt.max() == 0:
            print("  â˜¢â€…GT appears all-zero.")
        gt = (gt>0).astype(np.uint8)*255
    else:
        print("  â˜¢â€…No GT file found for this image:", ann_path)

    candidates = generate_candidates(img)
    print(f"  {len(candidates)} candidate masks generated.")

    # prepare candidate debug folder
    cand_dir = output_dir_p / f"{img_path.stem}_candidates"
    cand_dir.mkdir(parents=True, exist_ok=True)
    # save preprocessed image for inspection
    try:
        pre_clahe = apply_clahe(img, clip=2.0, tile=(8,8))
        Image.fromarray(pre_clahe).save(cand_dir / "pre_clahe.png")
    except Exception as e:
        print("  Could not save CLAHE preview:", e)

    best_iou = -1.0
    best_dice = 0.0
    best_mask = None
    best_method = None

    for idx, (name, params, mask) in enumerate(candidates):
        mask_u = (mask > 0).astype(np.uint8) * 255

        # Clean every candidate (aggressive for known noisy ones)
        if img_path.stem.lower() in {"training5", "training9"}:
            mask_u = clean_mask(mask_u, min_size=600, area_percentile=10, ecc_max=0.95,
                                opening_k=7, closing_k=13, median_ksize=7, take_convex_hull=False)
        else:
            mask_u = clean_mask(mask_u, min_size=300, area_percentile=20, ecc_max=0.96,
                                opening_k=5, closing_k=9, median_ksize=5, take_convex_hull=False)

        # Smart selection: remove outside / pick components likely to be cysts
        sel_mask = select_components(mask_u, img,
                                     expected_min_area=EXPECTED_MIN_AREA,
                                     expected_max_area=EXPECTED_MAX_AREA,
                                     center_y_range=EXPECTED_CENTER_Y_RANGE)

        # Save cleaned candidate and selected components for inspection
        try:
            Image.fromarray(mask_u).save(cand_dir / f"cleaned_{idx:02d}_{name}.png")
            Image.fromarray(sel_mask).save(cand_dir / f"selected_{idx:02d}_{name}.png")
        except Exception:
            pass

        if gt is not None:
            iou, dice = compute_metrics(sel_mask, gt)
            print(f"   cand {idx:02d}: {name}, params={params}, area={(sel_mask>0).sum()}, IoU={iou:.4f}, Dice={dice:.4f})")
            if iou > best_iou:
                best_iou, best_dice, best_mask, best_method = iou, dice, sel_mask.copy(), (name, params)
            if best_iou >= EARLY_STOP_IoU:
                print(f"  âœ… Early stop: IoU {best_iou:.4f} using {best_method[0]}")
                break
        else:
            area = (sel_mask>0).sum()
            if best_mask is None or area > (best_mask>0).sum():
                best_mask = sel_mask.copy()
                best_method = (name, params)

    # If no good mask selected, try permissive fallback attempts (helps TRAINING9)
    if best_mask is None or (best_mask>0).sum() == 0:
        print("  âž’ No good mask selected. Trying permissive fallback attempts.")
        # fallback 1: Otsu on CLAHE without inverting, larger structuring elements
        fb1 = strat_otsu_clahe(img, invert=False, clahe_clip=1.2, blur_k=1, open_k=3, min_size=20, min_hole=20)
        fb1 = clean_mask(fb1, min_size=50, area_percentile=1, ecc_max=0.99, opening_k=3, closing_k=5, median_ksize=3)
        sel1 = select_components(fb1, img, expected_min_area=20, expected_max_area=EXPECTED_MAX_AREA, center_y_range=EXPECTED_CENTER_Y_RANGE)

        # fallback 2: adaptive threshold more permissive
        fb2 = strat_adaptive(img, invert=False, blur_k=1, blockSize=15, C=5, min_size=20, min_hole=20)
        fb2 = clean_mask(fb2, min_size=50, area_percentile=1, ecc_max=0.99, opening_k=3, closing_k=5, median_ksize=3)
        sel2 = select_components(fb2, img, expected_min_area=20, expected_max_area=EXPECTED_MAX_AREA, center_y_range=EXPECTED_CENTER_Y_RANGE)

        # choose the larger area of sel1/sel2
        if (sel1>0).sum() >= (sel2>0).sum() and (sel1>0).sum() > 0:
            best_mask = sel1
            best_method = ('fallback1', {})
        elif (sel2>0).sum() > 0:
            best_mask = sel2
            best_method = ('fallback2', {})
        else:
            # final resort: keep largest cleaned original candidate (without select)
            possible = []
            for idx, (name, params, mask) in enumerate(candidates):
                mask_u = (mask > 0).astype(np.uint8) * 255
                mask_u = clean_mask(mask_u, min_size=50, area_percentile=1, ecc_max=0.99,
                                    opening_k=3, closing_k=5, median_ksize=3)
                possible.append(((mask_u>0).sum(), mask_u, (name, params))) # Changed to tuple for correct append
            if possible:
                # pick mask with largest area
                possible_sorted = sorted(possible, key=lambda x: x[0], reverse=True)
                best_mask = possible_sorted[0][1]
                best_method = ('fallback_largest', possible_sorted[0][2])
            else:
                best_mask = np.zeros_like(img, dtype=np.uint8)
                best_method = ('none', {})

        # save fallback mask
        try:
            Image.fromarray(best_mask).save(output_dir_p / f"{img_path.stem}_fallback_mask.png")
        except Exception:
            pass

    overlay, ncont = overlay_contours(img, best_mask)
    save_uint8(output_dir_p / f"{img_path.stem}_best_mask.png", best_mask)
    Image.fromarray(overlay).save(output_dir_p / f"{img_path.stem}_best_overlay.png")

    row = {"image": img_path.name, "contours_found": int(ncont), "best_method": str(best_method)}
    if gt is not None:
        row.update({"IoU": float(best_iou), "Dice": float(best_dice)})
        print(f"  â†’ Method: {best_method[0]}, IoU: {best_iou:.4f}, Dice: {best_dice:.4f}")
    else:
        print(f"  No GT â€” saved mask from {best_method[0]}")
    rows.append(row)

    # Per-image visualization (optional)
    try:
        fig, axs = plt.subplots(1, 3, figsize=(12,4))
        axs[0].imshow(img, cmap='gray'); axs[0].set_title("Original"); axs[0].axis("off")
        axs[1].imshow(best_mask, cmap='gray'); axs[1].set_title("Predicted Mask"); axs[1].axis("off")
        axs[2].imshow(overlay); axs[2].set_title("Overlay"); axs[2].axis("off")
        plt.suptitle(f"{img_path.name}\nMethod: {best_method[0]} | IoU: {best_iou:.3f} | Dice: {best_dice:.3f}")
        plt.tight_layout(); plt.show()
    except Exception:
        pass

# --------------------------
# Save & summary and bar chart
# --------------------------
df = pd.DataFrame(rows)
df.to_csv(output_dir_p / "metrics_visualized.csv", index=False)
print("\nâœ… Saved all results to:", output_dir_p)

print("\n--- Metrics dataframe head and dtypes ---")
print(df.head())
print(df.dtypes)

if "IoU" in df.columns:
    df["IoU"] = pd.to_numeric(df["IoU"], errors="coerce")
if "Dice" in df.columns:
    df["Dice"] = pd.to_numeric(df["Dice"], errors="coerce")

df_plot = df.copy().reset_index(drop=True)

# Added a conditional check for column existence before calling dropna
if "IoU" in df_plot.columns and "Dice" in df_plot.columns:
    df_plot = df_plot.dropna(subset=["IoU", "Dice"], how="any").reset_index(drop=True)

if df_plot.shape[0] == 0:
    print("âš ï¸ No numeric IoU/Dice values to plot.")
else:
    mean_iou = df_plot["IoU"].mean()
    mean_dice = df_plot["Dice"].mean()
    print(f"\nðŸ“Š Mean IoU: {mean_iou:.4f}")
    print(f"ðŸ“Š Mean Dice: {mean_dice:.4f}")

    n = len(df_plot)
    x = np.arange(n)
    width = 0.35

    plt.figure(figsize=(max(8, n * 0.6), 6))
    bars1 = plt.bar(x - width/2, df_plot["IoU"].values, width, label="IoU")
    bars2 = plt.bar(x + width/2, df_plot["Dice"].values, width, label="Dice")
    plt.ylim(0, 1.02)

    if n <= 20:
        plt.xticks(x, df_plot["image"], rotation=45, ha='right', fontsize=9)
    else:
        k = max(1, n // 20)
        ticks = x[::k]
        labels = df_plot["image"].tolist()
        plt.xticks(ticks, [labels[i] for i in ticks], rotation=45, ha='right', fontsize=9)

    plt.ylabel("Score")
    plt.title(f"IoU & Dice per Image (Mean IoU={mean_iou:.3f}, Mean Dice={mean_dice:.3f})")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.legend()

    def annotate_bars(bars, fmt="{:.3f}"):
        for b in bars:
            h = b.get_height()
            plt.gca().annotate(fmt.format(h),
                               xy=(b.get_x() + b.get_width() / 2, h),
                               xytext=(0, 4),
                               textcoords="offset points",
                               ha="center", va="bottom", fontsize=8)

    annotate_bars(bars1)
    annotate_bars(bars2)
    plt.tight_layout()
    plt.show()

    savepath = output_dir_p / "iou_dice_bar_chart.png"
    plt.savefig(savepath, dpi=200, bbox_inches='tight')
    print(f"Saved bar chart to: {savepath}")

# Post-run low IoU summary
if "IoU" in df.columns:
    print("\n--- Summary rows with NaN or IoU < 0.05 ---")
    low = df[df["IoU"].isna() | (df["IoU"] < 0.05)]
    if low.shape[0] > 0:
        print(low.to_string(index=True))
    else:
        print("No very low / NaN IoU rows found.")
else:
    print("âš ï¸ No IoU column in DataFrame â€” likely no GT masks were present.")
```
